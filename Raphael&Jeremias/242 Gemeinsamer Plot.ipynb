```python
%matplotlib inline
import matplotlib
import numpy as np
matplotlib.rcParams['figure.figsize'] = (12,8)
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy

plt.yscale('log')
plt.xscale('log')

x = np.array([0.1,0.3,0.6,1])
x2 = np.array([3])
x3 = np.array([6,10,30,60,100,200,300])
y = np.array([5.3,5.26,5.22,5.14])/0.03
y2 = np.array([4.22])/0.03
y3 = np.array([2.92,1.92,0.7,0.35,0.21,0.11,0.07])/0.03

yerr = y/20
y2err = y2/20
y3err = y3/20

fit = lambda x,m,c: m*x+c

logx = np.log10(x)
logy = np.log10(y)

logy3 = np.log10(y3)
logx3 = np.log10(x3)


popt, pcov = curve_fit(fit, logx, logy)
m = popt[0]
c = popt[1]

print('m = {:}, c = {:}+-{:}'.format(m, 10**c,10**pcov[1][1]))
print()
line = fit(logx, popt[0] ,popt[1])
linline = 10**line
plt.errorbar(x,y, yerr = yerr,color = 'green', ls = 'none', marker = '.')
plt.plot(x,linline, color='green', label='Diagr. 5')

plt.errorbar(x2,y2, yerr = y2err,color = 'green', ls = 'none', marker = '.')

popt, pcov = curve_fit(fit, logx3, logy3)

m = popt[0]
c = popt[1]


# Make a line\n",
line = fit(logx3, m ,c)

# Recall that this line is still in log space, so transform it back\n",
linline = 10**line
xline = 10**logx3

plt.plot(x3, linline, color= 'green', lw=1)

plt.errorbar(x3,y3, yerr = y3err,color = 'green', ls = 'none', marker = '.')
plt.plot(x3, fit(x3, *popt), color='green')
###########################################################################
U1=0.03
x = np.array([0.1,0.3,0.6,1,3])
x2 = np.array([6,10])
x3 = np.array([30,60,100,200,300])
y = np.array([2.02,2.02,2.0,2.0,1.96])/U1
y2 = np.array([1.76, 1.45])/U1
y3 = np.array([0.66,0.35,0.21,0.11,0.07])/U1

yerr = y/20
y2err = y2/20
y3err = y3/20

fit = lambda x,m,c: m*x+c

logx = np.log10(x)
logy = np.log10(y)

logy3 = np.log10(y3)
logx3 = np.log10(x3)


popt, pcov = curve_fit(fit, logx, logy)
m = popt[0]
c = popt[1]

print('m = {:}, c = {:}+-{:}'.format(m, 10**c,10**pcov[1][1]))
line = fit(logx, popt[0] ,popt[1])
linline = 10**line
plt.errorbar(x,y, yerr = yerr,color = 'yellow', ls = 'none', marker = '.')
plt.plot(x,linline, color='yellow', label='Diagr. 6')

plt.errorbar(x2,y2, yerr = y2err,color = 'yellow', ls = 'none', marker = '.')

popt, pcov = curve_fit(fit, logx3, logy3)

m = popt[0]
c = popt[1]


# Make a line\n",
line = fit(logx3, m ,c)

# Recall that this line is still in log space, so transform it back\n",
linline = 10**line
xline = 10**logx3

plt.plot(x3, linline, color= 'yellow', lw=1)

plt.errorbar(x3,y3, yerr = y3err,color = 'yellow', ls = 'none', marker = '.')
plt.plot(x3, fit(x3, *popt), color='yellow')
print(pcov, popt)


###########################################################################
U1 = 0.1
x = np.array([0.1,0.3,0.6,1,3,6,10])
x2 = np.array([30,60])
x3 = np.array([100,200,300])
y = np.array([1.20,1.2,1.2,1.2,1.2,1.19,1.18])/U1
y2 = np.array([1.05,0.83])/U1
y3 = np.array([0.6,0.33,0.23])/U1

yerr = y/20
y2err = y2/20
y3err = y3/20


fit = lambda x,m,c: m*x+c

logx = np.log10(x)
logy = np.log10(y)

logy3 = np.log10(y3)
logx3 = np.log10(x3)


popt, pcov = curve_fit(fit, logx, logy)
m = popt[0]
c = popt[1]

print('m = {:}, c = {:}+-{:}'.format(m, 10**c,10**pcov[1][1]))
print()
line = fit(logx, popt[0] ,popt[1])
linline = 10**line
plt.errorbar(x,y, yerr = yerr,color = 'cyan', ls = 'none', marker = '.')
plt.plot(x,linline, color='cyan', label='Diagr. 7')

plt.errorbar(x2,y2, yerr = y2err,color = 'cyan', ls = 'none', marker = '.')

popt, pcov = curve_fit(fit, logx3, logy3)

m = popt[0]
c = popt[1]


# Make a line\n",
line = fit(logx3, m ,c)

# Recall that this line is still in log space, so transform it back\n",
linline = 10**line
xline = 10**logx3

plt.plot(x3, linline, color= 'cyan', lw=1)

plt.errorbar(x3,y3, yerr = y3err,color = 'cyan', ls = 'none', marker = '.')
plt.plot(x3, fit(x3, *popt), color='cyan')
print(pcov, popt)


###########################################################################
U1 = 0.1
x = np.array([0.1,0.3,0.6,1])
x2 = np.array([3,6])
x3 = np.array([10,30,60,100,200,300])
y = np.array([1.2,1.2,1.19,1.18])/U1
y2 = np.array([1.06,0.86])/U1
y3 = np.array([0.63,0.24,0.12,0.08,0.04,0.03])/U1

yerr = y/20
y2err = y2/20
y3err = y3/20


fit = lambda x,m,c: m*x+c

logx = np.log10(x)
logy = np.log10(y)

logy3 = np.log10(y3)
logx3 = np.log10(x3)


popt, pcov = curve_fit(fit, logx, logy)
m = popt[0]
c = popt[1]

print('m = {:}, c = {:}'.format(m, 10**c))
print()
line = fit(logx, popt[0] ,popt[1])
linline = 10**line
plt.errorbar(x,y, yerr = yerr,color = 'grey', ls = 'none', marker = '.')
plt.plot(x,linline, color='grey', label='Diagr. 8')

plt.errorbar(x2,y2, yerr = y2err,color = 'grey', ls = 'none', marker = '.')

popt, pcov = curve_fit(fit, logx3, logy3)

m = popt[0]
c = popt[1]


# Make a line\n",
line = fit(logx3, m ,c)

# Recall that this line is still in log space, so transform it back\n",
linline = 10**line
xline = 10**logx3

plt.plot(x3, linline, color= 'grey', lw=1)

plt.errorbar(x3,y3, yerr = y3err,color = 'grey', ls = 'none', marker = '.')
plt.plot(x3, fit(x3, *popt), color='grey')
print(pcov, popt)


###########################################################################
plt.title('Diagramm 10: Frequenzgänge aus Diagr.5-8 ', fontsize='14')
plt.xlabel(r'Frequenz $\nu [kHz]$', fontsize='17')
plt.ylabel(r'Betriebsverstärkung $V$´ $[1]$', fontsize='17')
plt.legend(loc='lower left', fontsize='14')
plt.savefig("242 2c gemeinsamer plot.pdf", format='pdf', dpi = '1000')
```
