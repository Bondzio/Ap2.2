```python
plt.yscale('log')
plt.xscale('log')

x = np.array([0.1,0.3,0.6,1])
x2 = np.array([3])
x3 = np.array([6,10,30,60,100,200,300])
y = np.array([5.3,5.26,5.22,5.14])/0.03
y2 = np.array([4.22])/0.03
y3 = np.array([2.92,1.92,0.7,0.35,0.21,0.11,0.07])/0.03

yerr = y/20
y2err = y2/20
y3err = y3/20

fit = lambda x,m,c: m*x+c

logx = np.log10(x)
logy = np.log10(y)

logy3 = np.log10(y3)
logx3 = np.log10(x3)


popt, pcov = curve_fit(fit, logx, logy)
m = popt[0]
c = popt[1]

print('m = {:}, c = {:}+-{:}'.format(m, 10**c,10**pcov[1][1]))
print()
line = fit(logx, popt[0] ,popt[1])
linline = 10**line
plt.errorbar(x,y, yerr = yerr,color = 'green', ls = 'none', marker = '.')
plt.plot(x,linline, color='green', label=''.join("Steigung $m_1$ = {:.2f} $\pm$ {:.2f}".format(popt[0], np.sqrt(pcov[0,0]))))

plt.errorbar(x2,y2, yerr = y2err,color = 'black', ls = 'none', marker = '.')

popt, pcov = curve_fit(fit, logx3, logy3)

m = popt[0]
c = popt[1]


# Make a line\n",
line = fit(logx3, m ,c)

# Recall that this line is still in log space, so transform it back\n",
linline = 10**line
xline = 10**logx3

plt.plot(x3, linline, color= 'blue', lw=1)

plt.errorbar(x3,y3, yerr = y3err,color = 'black', ls = 'none', marker = '.')
plt.plot(x3, fit(x3, *popt), color='blue', label=''.join("Steigung $m_1$ = {:.2f} $\pm$ {:.2f}".format(popt[0], np.sqrt(pcov[0,0]))))

plt.title('Diagramm 5: Frequenzgang ($R _{G1}$) ', fontsize='14')
plt.xlabel(r'Frequenz $\nu [Hz]$', fontsize='17')
plt.ylabel(r'Betriebsverstärkung $V$´ $[1]$', fontsize='17')
plt.legend(loc='upper right', fontsize='14')
#plt.text(-5.9,620, "n-Werte der gemessenen Maxima: {:.2f}$\pm$0.03, {:.2f}$\pm$0.03, {:.2f}$\pm$0.03, {:.2f}$\pm$0.03, {:.2f}$\pm$0.03, {:.2f}$\pm$0.03, {:.2f}$\pm$0.03, {:.2f}$\pm$0.03, {:.2f}$\pm$0.03 \nn_theo = -4.48, -3.47, -2.46, -1.43, 0, 1.43, 2.46, 3.47, 4.48".format(line(y2)[0],line(y2)[1],line(y2)[2],line(y2)[3],line(y2)[4],line(y2)[5],line(y2)[6],line(y2)[7],line(y2)[8]))

#print(line(y2))
plt.savefig("242 2a RG1.pdf", format='pdf', dpi = '1000')
```
