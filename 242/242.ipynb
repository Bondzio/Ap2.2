```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
#Aufgabe 1a
Ue  = np.array([200,150,100,50,-50, -100, -150, -200])/1000
dUe = 0.03*Ue
Ua1 = np.array([-3.16, -2.48, -1.66, -0.82, 0.78, 1.54, 2.32, 3.16])
dUa1= 0.03*Ua1
Ua2 = np.array([-12.9, -12.9, -9.1, -4.6, 4.6, 8.9, 13.1, 13.9])
dUa2= 0.03*Ua2

Re  = 3000
dRe = 0.05*Re
Rg1 = 48.7*1000
dRg1 = 0.05*Rg1
Rg2 = 274*1000
dRg2 = 0.05*Rg2

V1_theo = Rg1/Re
dV1_theo= V1_theo*np.sqrt((dRg1/Rg1)**2+(dRe/Re)**2)
V2_theo = Rg2/Re
dV2_theo= V2_theo*np.sqrt((dRg2/Rg2)**2+(dRe/Re)**2)
print('V1_theo', V1_theo, '+-', dV1_theo)
print('V2_theo', V2_theo, '+-', dV2_theo)

#Fit
def fit_func(ue, v):
    return -ue*v

popt, pcov=curve_fit(fit_func,Ue,Ua1,p0=[1],sigma=dUa1)
print("V1_ex=",popt[0],",Standardfehler=",np.sqrt(pcov[0][0]))
plt.plot(Ue,fit_func(Ue,*popt), label='Fit')
print('Abweichung', (V1_theo-popt[0])/(np.sqrt((pcov[0][0])+(dV1_theo)**2)))
print("chi2_red=",np.sum((fit_func(Ue,*popt)-Ua1)**2/dUa1**2)/(len(Ua1)-1))


popt, pcov=curve_fit(fit_func,Ue[2:Ue.size-2],Ua2[2:Ue.size-2],p0=[1],sigma=dUa2[2:Ue.size-2])
print("V2_ex=",popt[0],",Standardfehler=",np.sqrt(pcov[0][0]))
plt.plot(Ue,fit_func(Ue,*popt), label='Fit')
print('Abweichung', (V2_theo-popt[0])/(np.sqrt((pcov[0][0])+(dV2_theo)**2)))
print("chi2_red=",np.sum((fit_func(Ue[2:Ue.size-2],*popt)-Ua2[2:Ue.size-2])**2/dUa2[2:Ue.size-2]**2)/(len(Ua2[2:Ue.size-2])-1))

#Plot
plt.title('Abb.1:Verstärkung Gleichspannungsverstärker')
plt.xlabel('$U_{ein}$ in V')
plt.ylabel('$U_{aus}$ in V')
plt.errorbar(Ue, Ua1, yerr=dUa1, xerr=dUe, linestyle='None', fmt='.', label='$R=48.7k\\Omega$')
plt.errorbar(Ue, Ua2, yerr=dUa2, xerr=dUe, linestyle='None', fmt='.', label='$R=274k\\Omega$')
plt.ylim([-16,16])
plt.xlim([-0.22,0.22])
plt.legend(loc='best',prop={'size':10}) #Legendengroesse
plt.savefig('figures/Gleichspannungsverstärker.pdf',format='pdf')


```

```python
#Aufgabe 1b
Ue  = np.array([100, 200, 400, 600, 800, 1000])/10000
dUe = 20/10000
Ua1 = np.array([0.96, 1.6, 3.12, 4.56, 5.84, 7.28])
dUa1= 0.2
Ua2 = np.array([2.6, 4.2, 7.8, 11.4, 15.0, 18.4])
dUa2= 0.2

Re  = 3000
dRe = 0.05*Re
Rg1 = 274*1000
dRg1= 0.05*Rg1
Rg2 = 680*1000
dRg2=0.05*Rg2

V1_theo = Rg1/Re
dV1_theo= V1_theo*np.sqrt((dRg1/Rg1)**2+(dRe/Re)**2)
V2_theo = Rg2/Re
dV2_theo= V2_theo*np.sqrt((dRg2/Rg2)**2+(dRe/Re)**2)
print('V1_theo', V1_theo, '+-', dV1_theo)
print('V2_theo', V2_theo, '+-', dV2_theo)

#Fit
def fit_func(ue, v):
    return ue*v

popt, pcov=curve_fit(fit_func,Ue,Ua1,p0=[1],sigma=dUa1)
print("V1_ex=",popt[0],",Standardfehler=",np.sqrt(pcov[0][0]))
print('Abweichung', (V1_theo-popt[0])/(np.sqrt((pcov[0][0])+(dV1_theo)**2)))
plt.plot(Ue,fit_func(Ue,*popt), label='Fit')
print("chi2_red=",np.sum((fit_func(Ue,*popt)-Ua1)**2/dUa1**2)/(len(Ua1)-1))

popt, pcov=curve_fit(fit_func,Ue,Ua2,p0=[1],sigma=dUa2)
print("V2_ex=",popt[0],",Standardfehler=",np.sqrt(pcov[0][0]))
print('Abweichung', (V2_theo-popt[0])/(np.sqrt((pcov[0][0])+(dV2_theo)**2)))
plt.plot(Ue,fit_func(Ue,*popt), label='Fit')
print("chi2_red=",np.sum((fit_func(Ue,*popt)-Ua2)**2/dUa2**2)/(len(Ua2)-1))

#Plot
plt.title('Abb.2:Verstärkung Wechselspannungsverstärker')
plt.xlabel('$U_{ein}$ in V')
plt.ylabel('$U_{aus}$ in V')
plt.errorbar(Ue, Ua1, yerr=dUa1, xerr=dUe, linestyle='None', fmt='.', label='$R=274k\\Omega$')
plt.errorbar(Ue, Ua2, yerr=dUa2, xerr=dUe, linestyle='None', fmt='.', label='$R=680k\\Omega$')
plt.ylim([0,21])
plt.xlim([0,0.11])
plt.legend(loc='best',prop={'size':10}) #Legendengroesse
plt.savefig('figures/Wechselspannungsverstärker.pdf',format='pdf')

```
