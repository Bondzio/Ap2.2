```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Teil 1: Absorption von $\beta$-Strahlung in Aluminium
--

```python
#Index 1 steht für Versuchsreihe "Absorption von beta-Strahlung in Aluminium"
null = 101/(5*60)
error_null = np.sqrt(101)/(5*60)

#Das letzte Element wird gelöscht, da es genau dem Nulleffekt entspricht. Bei Logarithmischen Plots führt dies zu Problemen
counts1 = np.array([1156, 660, 406, 287, 788, 467, 292, 205, 110, 83, 154])
#counts1 = counts1[0:counts1.size-1]
time1 = np.array([30, 30, 30, 30, 120, 120, 120, 120, 120, 120, 300])
#time1 = time1[0:time1.size-1]
thickness1_mm = np.array([0, 0.3, 0.6, 0.9, 1.2, 1.5, 1.8, 2.1, 2.4, 2.7, 3.7]) #in mm
thickness1 = thickness1_mm/10 # in cm
#thickness1 = thickness1[0:thickness1.size-1]

cps1 = counts1/time1-null #(n-n_0^Beta)
error_cps1 = np.sqrt((np.sqrt(counts1)/time1)**2+(error_null)**2)

plt.yscale('log')
plt.title('Abb. 2: Absorption von $\\beta$-Strahlung in Aluminium')
plt.xlabel('Aluminiumdicke in cm')
plt.ylabel('Aktivität in counts/s')
plt.ylim((0.0387,100))
plt.errorbar(thickness1, cps1, yerr=error_cps1, linestyle='None', fmt='o')
plt.savefig('figures/Absorption_beta.pdf',format='pdf')

xmax = 0.34 #Von Hand, wie kriegen wir diesen Wert sauber ? Maximale Reichweite von Beta-Strahlung in Aluminium aus Diagramm
error_xmax = 0.04#Die Absorberdicke, ab der die Absorptionskurve nahezu senkrecht verlaufen würde

rho_al = 2.7 # g/cm^3 Aus Abbildung 8 im Praktikums-Skript bestimmt !
R_beta = xmax * rho_al #Flächendichte
error_R_beta = error_xmax * rho_al 
R_beta_ges = R_beta + 0.13 #Addiert die Fensterdicke der Präparatkapsel hinzu, so lassen.

print("R_beta=", R_beta, "+-", error_R_beta)
print("R_beta_ges=", R_beta_ges, "+-", error_R_beta)
#Damit erhält man eine Maximalenergie von 2.25+-0.2 MeV, abgelesen aus Abb 8 
```

Teil 2: Absorption von $\gamma$-Strahlung in Blei
--

```python
#Index 2 steht für Versuchsreihe "Absorption von gamma-Strahlung in Blei"

counts2 = np.array([4159, 2862, 1987, 1464, 1217, 873, 651, 462, 394, 315, 236])
thickness2 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])

cps2 = counts2/60-null #Wieder n-n_0 in Einheiten [s^-1]
error_cps2 = np.sqrt((np.sqrt(counts2)/60)**2+(error_null)**2)

#Fit
def fit_func(x,A,mu):
    return A*np.exp(-x*mu)
from scipy.optimize import curve_fit
popt, pcov=curve_fit(fit_func,thickness2,cps2,p0=[500,0.02],sigma=error_cps2)

plt.yscale('log')
plt.title('Abb.3: Absorption von $\\gamma$-Strahlung in Blei')
plt.xlabel('Bleidicke in cm')
plt.ylabel('Aktivität in counts/s')
plt.errorbar(thickness2, cps2, error_cps2, linestyle='None', fmt='o')
plt.plot(thickness2,fit_func(thickness2,*popt))
plt.savefig('figures/Absorption1_gamma.pdf',format='pdf')

chi2_= np.sum((fit_func(thickness2,*popt)-cps2)**2/error_cps2**2)
dof = len(cps2)-2 #degrees of freedom
chi2_red=chi2_/dof
print("chi2=",chi2_)
print("chi2_red=",chi2_red)

#Aus Steigung der Geraden ergibt sich der Schwächungskoeffizient mu, 
print("A=",popt[0],",Standardfehler=",np.sqrt(pcov[0][0]))
print("mu=",popt[1],",Standardfehler=",np.sqrt(pcov[1][1]))

mu = popt[1]
rho = 11.342 #in g/cm^3 

print("Massenschwächungskoeffizient:", "mu/rho=", mu/rho, "+-", np.sqrt(pcov[1][1])/rho)
print("Energie der gamma-Quanten:", "1.4 MeV")#Aus Abbildung 9 im Praktikumsskript abzulesen !
```

Teil 3: Bestimmung der Aktivität des $\gamma$-Strahlers
--

```python
#Index 3 steht für Versuchsreihe "Aktivität des gamma-Strahlers"
#alle Angaben in cm

counts3 = np.array([37950, 12386, 3569])
abstand3 = np.array([5, 10, 20])

cps3 = counts3/60-null#n-n_0
error_cps3 = np.sqrt((np.sqrt(counts3)/60)**2+(error_null)**2)

epsilon = 0.04  # Korrekturfaktor 4%
r = 0.7  #Radius Rohr
A = 4*cps3*abstand3**2/(epsilon*r**2) #Aktivität des Strahlers
l = 4 #Länge
k1 = (1+l/(2*abstand3))**2 #erster Korrekturfaktor
Akorr = A*k1#Dies ist die Raumwinkelkorrektur: Nicht alle Strahlung kommt am Detektor an, sodass Aktivität unterschätzt würde.
#Wählt man Omega als den minimalen Raumwinkel, so überschätzt man die Aktivität
#Hier wird also der übeschätzte Bereich durch den unterschätzten Bereich kompensiert durch Korrekturfaktor k1
plt.yscale('log')
plt.xscale('log')
plt.title(' Aktivität des $\\gamma$-Strahlers')
plt.xlabel('Abstand in cm')
plt.ylabel('Aktivität in counts/s')
plt.errorbar(abstand3, cps3, yerr=error_cps3, xerr=0.4, linestyle='None', fmt='.')
plt.savefig('figures/Aktivität.pdf',format='pdf')




x = 0.14 #Eindringtiefe, Skript [cm]
mu = 0.0526*7.9 #*dichte, Siehe Abb 9 im Skript für Bestimmung von mu/rho, siehe VII.3.2
k2 = np.exp(-mu*x)#Zweíter Korrekturfaktor
Aab = Akorr/k2#Korrektur der Absoprtion der Präperatkapsel an sich
d = abstand3


error_A = A*np.sqrt((2/d*0.4)**2+(epsilon*0.01)**2)
error_Akorr = Akorr*np.sqrt((2/(d+l/2)*0.4)**2+(1/epsilon*0.01)**2)
error_Aab = Aab*np.sqrt((2/(d+l/2)*0.4)**2+(1/epsilon*0.01)**2+(x*0.003)**2)

print(A/2)
print(error_A/2)

print(Akorr/2)#Mit ersten Korrekturfaktor für Raumwinkel
print(error_Akorr/2)
#Mit zweitem Korrekturfaktor für Absorption der Präparatkapsel
print(Aab/2)
print(error_Aab/2)
```

Teil 4: Absorptionsmessung und Energiebestimmung von $\alpha$-Strahlung
--

```python
#Index 4 steht für Versuchsreihe "Absorptionsmessung und Energiebestimmung von $\\alpha$-Strahlung"

pressure4 =  np.array([20, 121, 219, 269, 295, 322, 350, 334, 381, 368, 357, 397, 420, 470, 606, 375])#Aus Druckinitial,[mbar]
counts4 = np.array([4885, 4811, 4685, 4667, 4439, 3943, 2651, 3576, 958, 1611, 2269, 443, 139, 68, 61, 1328])

cps4 = counts4/60-null #n-n_0
error_cps4 = np.sqrt((np.sqrt(counts4)/60)**2+(error_null)**2)

#Fit
def fit_func(x,L,x0,k,c):
    return c+L/(1+np.exp(k*(x-x0)))
from scipy.optimize import curve_fit
popt, pcov=curve_fit(fit_func,pressure4,cps4,p0=[80,400,0.1,0],sigma=error_cps4)

#plt.yscale('log')
plt.title('Abb.4: Absorptionsmessung von $\\alpha$-Strahlung')
plt.xlabel('Druck in mbar')
plt.ylabel('Aktivität in counts/s')
plt.errorbar(pressure4, cps4, error_cps4, linestyle='None',fmt='.')
t=np.arange(0,700,0.1)
plt.plot(t,fit_func(t,*popt))
plt.savefig('figures/Absorption_alpha.pdf',format='pdf')


chi2_= np.sum((fit_func(pressure4,*popt)-cps4)**2/error_cps4**2)
dof = len(cps4)-4 #degrees of freedom
chi2_red=chi2_/dof
print("chi2=",chi2_)
print("chi2_red=",chi2_red)

print("L=",popt[0])
print("x0=",popt[1])
print("k=",popt[2])
print("c=",popt[3])

p_half = 355 #mit mathematica bestimmt, Druck, bei dem die Aktivität auf die Hälfte abgefallen ist
#Aus Tabelle 4 extrapolieren. Hälfte des Countspitzenwertes nehmen und daraus P1/2 extrapolieren. Geht auch aus unterem Plot

s0 = 4.45 #Abstand Präparat - Zählrohr aus Skript
p0 = 1013 #Normaldruck Skript
rho_gl = 2.35 #[mg/cm^2] Flächendichte des Glimmers, Korrektur um diesen
s2 = rho_gl / 2.35
s3 = 0.68 #Skript
s1 = p_half/p0*s0 + s2 + s3
print("Gesamtreichweite: ",s1, "cm")
#Daraus folgt eine Energie der alpha-Teilchen von 4.6 MeV, aus Abbildung 8
#Literaturwert ist E=5.48MeV aus Kript
```
