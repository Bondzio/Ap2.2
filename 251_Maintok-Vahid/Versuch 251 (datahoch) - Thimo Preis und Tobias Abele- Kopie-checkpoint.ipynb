```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importieren und grafische Darstellung der Messwerte
```

```python
#Zaehlrohrspannung
U = np.arange(420, 580, 25)
dU = 10
#gez채hlte Ereignisse
N = np.array([1838, 2242, 2296, 2309, 2288, 2276, 2412])
dN = np.sqrt(N)
```

```python
plt.errorbar(U, N, yerr = dN, xerr=dU, fmt = '.')
plt.xlabel('Spannung / V')
plt.ylabel('Ereignisse')
plt.title('Z채hlrohrcharakteristik')
```

```python
from scipy.optimize import curve_fit
def linear(x, a, b):
    return a * x + b
popt, pcov = curve_fit(linear, U[1:], N[1:])
```

```python
plt.errorbar(U, N, dN, fmt = '.')
plt.xlabel('Spannung / V')
plt.ylabel('Ereignisse')
plt.xlim(400, 600)
plt.title('Abb. 2: Z채hlrohrstatistik')
plt.plot(U, linear(U, *popt))
plt.savefig('figures/aehlrohrstatistik.pdf', format = 'pdf', bbox_inches = 'tight')
```

```python
chi2_ = np.sum((linear(U[1:], *popt)-N[1:])**2/dN[1:]**2)
dof = len(U[1:]) - 2
chi2_red = chi2_/dof
print(chi2_red)
```

### Auswertung der Daten mit hoher mittlerer Ereigniszahl

```python
anzahl, haeufigkeit = np.loadtxt('data/datahoch.dat', unpack = True)
fehler = np.sqrt(haeufigkeit)
```

```python
plt.errorbar(anzahl, haeufigkeit, fehler, fmt = '.')
plt.xlabel('Anzahl der Zerf채lle pro Zeiteinheit in 1/s')
plt.ylabel('H채ufigkeit')
plt.title('Statistik des radioaktiven Zerfalls')
#plt.yscale('log')
```

```python
from numpy import exp, pi, sqrt
def gaussian(x, A, mu, sig): # A: Fl채che der Gau-Funktion
    return A/sqrt(2*pi)/sig * exp(-(x - mu)**2/2/sig**2)
```

```python
popt, pcov = curve_fit(gaussian, anzahl[12:-20], haeufigkeit[12:-20], p0 = [2016, 75, 9], sigma = fehler[12:-20])
```

```python
from scipy.special import gamma
def poisson(x, A_p, mu_p):
    return A_p * exp(-mu_p) * mu_p**x / gamma(x+1)
popt_p, pcov_p = curve_fit(poisson, anzahl[12:-20], haeufigkeit[12:-20], p0 = [2000, 75], sigma = fehler[12:-20])
```

```python
plt.errorbar(anzahl, haeufigkeit, fehler, fmt = '.', label = 'Messung')
plt.xlabel('Anzahl der Zerf채lle pro Zeiteinheit in 1/s')
plt.ylabel('H채ufigkeit')
plt.title('Abb. 3: Statistik des radioaktiven Zerfalls')
x = np.linspace(40, 110, 100)
plt.plot(x, gaussian(x, *popt), label = 'Gauss')
plt.plot(x, poisson(x, *popt_p), label = 'Poisson', linestyle = ':')
plt.legend()
plt.savefig('figures/poisson_vs_gaussian_hoch1.pdf', format = 'pdf', bbox_inches = 'tight')
```

```python
print('Gaussfit:')
print('A = ', popt[0], ', Standardfehler = ', np.sqrt(pcov[0][0]))
print('mu = ', popt[1], ', Standardfehler = ', np.sqrt(pcov[1][1]))
print('sig = ', popt[2], ', Standardfehler = ', np.sqrt(pcov[2][2]))
print('Poissonfit: ')
print('A_p = ', popt_p[0], ', Standardfehler = ', sqrt(pcov_p[0][0]))
print('mu_p = ', popt_p[1], ', Standardfehler = ', sqrt(pcov_p[1][1]))
```

```python
# Gauss:
chi2_g = np.sum((gaussian(anzahl[12:-20], *popt) - haeufigkeit[12:-20])**2/fehler[12:-20]**2)
dof_g = len(anzahl[12:-20]) - 3 #dof: degrees of freedom, Freiheitsgrade
chi2_red_g = chi2_g / dof_g
print('chi2_g = ', chi2_g)
print('chi2_red_g = ', chi2_red_g)
#Poisson:
chi2_p = np.sum((poisson(anzahl[12:-20], *popt_p) - haeufigkeit[12:-20])**2/fehler[12:-20]**2)
dof_p = len(anzahl[12:-20]) - 2 # poisson hat nur 2 Parameter
chi2_red_p = chi2_p/dof_p
print('chi2_p = ', chi2_p)
print('chi2_red_p = ', chi2_red_p)
```

```python
from scipy.stats import chi2
# Gauss:
prob_g = round(1 - chi2.cdf(chi2_g, dof_g), 2) * 100
# Poisson:
prob_p = round(1 - chi2.cdf(chi2_p, dof_p), 2) * 100
print('Wahrscheinlichkeit Gauss = ', prob_g, '%')
print('Wahrscheinlichkeit Poisson = ', prob_p, '%')
```
