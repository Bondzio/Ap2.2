```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

Zu Teil 1b)
----

```python
Zaehlrate = np.loadtxt("data/15_05_2017 14_30_00.txt", usecols=(1,))
Winkel = np.loadtxt("data/15_05_2017 14_30_00.txt", usecols=(0,))
plt.plot(Winkel, Zaehlrate)
plt.title("Abb.4:Zählrate in Abhängigkeit des Einfallswinkels")
plt.xlabel("Winkel [°]")
plt.ylabel("Zaehlrate [s^-1]")
plt.savefig("figures/abb2ZaehlratevomWinkel.pdf")
```

Zu Aufgabenteil 1c)
--

```python
U=np.arange(20000,35000,1000)
cps = np.array([1.15,1.7,3.15,6.85,31.85,57.50,84.50,108.3,133.6,157.2,179.1,202.1,225.5,246.0,281.5])

def line(x,a, b):
    return a*x + b
popt, pcov = curve_fit(line, U, cps)

error_cps = np.sqrt(cps)
error_U = 1000
werte=np.linspace(20000,25000,1000)
y = werte-werte
#plt.yscale('log')
plt.title('Abb.3: Spannungsabhängigkeit der Röntgenstrahlung')
plt.xlabel('Spannung in V')
plt.ylabel('Aktivität in counts/s')
#plt.ylim((0,430))
plt.errorbar(U, cps, yerr=error_cps, linestyle='None', fmt='o')
plt.plot(U, line(U, popt[0], popt[1]), 'r-')
plt.plot(werte,y, color="orange")
plt.savefig('figures/Spannungsabhängigkeit.pdf',format='pdf')
```
