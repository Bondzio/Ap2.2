```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Index 1 steht für Versuchsreihe "Absorption von beta-Strahlung in Aluminium"
null = 135/(5*60)
error_null = np.sqrt(135)/(5*60)

#Das letzte Element wird gelöscht, da es genau dem Nulleffekt entspricht. Bei Logarithmischen Plots führt dies zu Problemen
counts1 = np.array([1455, 779, 508, 328, 253, 620, 378, 234, 152, 93, 77, 135])
#counts1 = counts1[0:counts1.size-1]
time1 = np.array([30, 30, 30, 30, 30, 120, 120, 120, 120, 120, 120, 300])
#time1 = time1[0:time1.size-1]
thickness1_mm = np.array([0, 0.3, 0.6, 0.9, 1.2, 1.5, 1.8, 2.1, 2.4, 2.7, 3.0, 4.0]) #in mm
thickness1 = thickness1_mm/10 # in cm
#thickness1 = thickness1[0:thickness1.size-1]

cps1 = counts1/time1-null
error_cps1 = np.sqrt((np.sqrt(counts1)/time1)**2+(error_null)**2)

plt.yscale('log')
plt.title('Absorption von $\\beta$-Strahlung in Aluminium')
plt.xlabel('Aluminiumdicke in cm')
plt.ylabel('Aktivität in counts/s')
plt.ylim((0.0387,100))
plt.errorbar(thickness1, cps1, yerr=error_cps1, linestyle='None', fmt='o')
plt.savefig('figures/Absorption_beta.pdf',format='pdf')

xmax = 0.34
error_xmax = 0.04

rho_al = 2.7 # g/cm^3
R_beta = xmax * rho_al
error_R_beta = error_xmax * rho_al
R_beta_ges = R_beta + 0.13

print("R_beta=", R_beta, "+-", error_R_beta)
print("R_beta_ges=", R_beta_ges, "+-", error_R_beta)
#Damit erhält man eine Maximalenergie von 2.25+-0.2 MeV
```

```python
#Index 2 steht für Versuchsreihe "Absorption von gamma-Strahlung in Blei"

counts2 = np.array([4278, 2870, 2158, 1665, 1269, 966, 686, 500, 409, 304, 279])
thickness2 = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5])

cps2 = counts2/60-null
error_cps2 = np.sqrt((np.sqrt(counts2)/60)**2+(error_null)**2)

#Fit
def fit_func(x,A,mu):
    return A*np.exp(-x*mu)
from scipy.optimize import curve_fit
popt, pcov=curve_fit(fit_func,thickness2,cps2,p0=[500,0.02],sigma=error_cps2)

plt.yscale('log')
plt.title('Absorption von $\\gamma$-Strahlung in Blei')
plt.xlabel('Bleidicke in cm')
plt.ylabel('Aktivität in counts/s')
plt.errorbar(thickness2, cps2, error_cps2, linestyle='None', fmt='o')
plt.plot(thickness2,fit_func(thickness2,*popt))
plt.savefig('figures/Absorption_gamma.pdf',format='pdf')


print("A=",popt[0],",Standardfehler=",np.sqrt(pcov[0][0]))
print("mu=",popt[1],",Standardfehler=",np.sqrt(pcov[1][1]))

mu = popt[1]
rho = 11.342 #in g/cm^3

print("Massenschwächungskoeffizient:", "mu/rho=", mu/rho, "+-", np.sqrt(pcov[1][1])/rho)
print("Energie der gamma-Quanten:", "1.4 MeV")
```

```python
#Index 3 steht für Versuchsreihe "Aktivität des gamma-Strahlers"
#alle Angaben in cm

counts3 = np.array([52892, 14777, 3903])
abstand3 = np.array([5, 10, 20])

cps3 = counts3/60-null
error_cps3 = np.sqrt((np.sqrt(counts3)/60)**2+(error_null)**2)

epsilon = 0.04
r = 0.7
A = 4*cps3*abstand3**2/(epsilon*r**2)
l = 4
k1 = (1+l/(2*abstand3))**2
Akorr = A*k1

plt.yscale('log')
plt.xscale('log')
plt.title('Aktivität des $\\gamma$-Strahlers')
plt.xlabel('Abstand in cm')
plt.ylabel('Aktivität in counts/s')
plt.errorbar(abstand3, cps3, yerr=error_cps3, xerr=0.4, linestyle='None', fmt='.')
plt.savefig('figures/Aktivität.pdf',format='pdf')


print(A)
print(Akorr)

x = 0.14
mu = 0.057*7.9 
k2 = np.exp(-mu*x)
Aab = Akorr/k2
d = abstand3

print(Aab)

error_Aab = Aab*np.sqrt((2/(d+l/2)*0.4)**2+(epsilon*0.01)**2+(x*0.003)**2)
print(error_Aab)
```

```python
#Index 4 steht für Versuchsreihe "Absorptionsmessung und Energiebestimmung von $\\alpha$-Strahlung"

pressure4 =  np.array([22, 124, 224, 324, 429, 534, 278, 304, 345, 362, 398, 419, 378, 655, 428, 440])
counts4 = np.array([4820, 4810, 4766, 3958, 80, 74, 4707, 4391, 2963, 2030, 423, 143, 1141, 83, 111, 86])

cps4 = counts4/60-null
error_cps4 = np.sqrt((np.sqrt(counts4)/60)**2+(error_null)**2)

#Fit
def fit_func(x,L,x0,k,c):
    return c+L/(1+np.exp(k*(x-x0)))
from scipy.optimize import curve_fit
popt, pcov=curve_fit(fit_func,pressure4,cps4,p0=[80,400,0.1,0],sigma=error_cps4)

#plt.yscale('log')
plt.title('Absorptionsmessung von $\\alpha$-Strahlung')
plt.xlabel('Druck in mbar')
plt.ylabel('Aktivität in counts/s')
plt.errorbar(pressure4, cps4, error_cps4, linestyle='None',fmt='.')
t=np.arange(0,700,0.1)
plt.plot(t,fit_func(t,*popt))
plt.savefig('figures/Absorption_alpha.pdf',format='pdf')


chi2_= np.sum((fit_func(pressure4,*popt)-cps4)**2/error_cps4**2)
dof = len(cps4)-4 #degrees of freedom
chi2_red=chi2_/dof
print("chi2=",chi2_)
print("chi2_red=",chi2_red)

print("L=",popt[0])
print("x0=",popt[1])
print("k=",popt[2])
print("c=",popt[3])

p_half = 355 #mit mathematica bestimmt, Druck, bei dem die Aktivität auf die Hälfte abgefallen ist

s0 = 4.45 #Abstand Präparat - Zählrohr
p0 = 1013 #Normaldruck
rho_gl = 2.35
s2 = rho_gl / 2.35
s3 = 0.68
s1 = p_half/p0*s0 + s2 + s3
print("Gesamtreichweite: ",s1, "cm")
#Daraus folgt eine Energie der alpha-Teilchen von 4.6 MeV
```
